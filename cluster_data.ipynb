```python
import numpy as np 
import pandas as pd 
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("CC GENERAL.csv")
data
```

```python
data["MINIMUM_PAYMENTS"]=data["MINIMUM_PAYMENTS"].fillna(data["MINIMUM_PAYMENTS"].mean() )
data["CREDIT_LIMIT"]=data["CREDIT_LIMIT"].fillna(data["CREDIT_LIMIT"].mean() )
data.isnull().sum()
```

```python
x=data.iloc[:,[1,3]]
x.head()
```

```python
means=KMeans(3)
means.fit(x)
```

```python
identified_clusters=means.fit_predict(x)
identified_clusters
```

```python
data_with_clust = data.copy()
data_with_clust['Clusters']=identified_clusters
plt.scatter(data_with_clust['BALANCE'],data_with_clust['PURCHASES'],c= data_with_clust['Clusters'],cmap='rainbow')
```
